```python
def tricky_list(lst):
    return lst[:-1] and lst[:-2] if len(lst[1]) == len(lst[-1]) else lst[1:]
```

```python
names = ['Ben', 'Ken', 'Len']
```

```python
print(tricky_list(names))
```
